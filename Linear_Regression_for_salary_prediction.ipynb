MODEL TRAINING AND TESTING

```python
from sklearn.model_selection import train_test_split
```

```python
x = data.drop("Salary" , axis = 1)  # dropping salary columns and storing it to x
```

```python
y = data['Salary']
y.head()
```

```python
# printing x and y
print(x.head())
print(y.head())
```

It is observed that x is now the years of experience and y is the salary

```python
xtrain , xtest , ytrain , ytest = train_test_split (x , y ,test_size = 0.2, random_state = 42)
# training 80 % of dataset and 20% for testing as given in test size
# also random state set to 42 describes that the data can go in any random order for training and testing

from sklearn.linear_model import LinearRegression
L = LinearRegression()
```

```python
L.fit(xtrain , ytrain)
```

```python
y_pred = L.predict(xtest)
print(L.score(xtest , ytest))
```

The score for the predicted value above shows that 89% accurately the model is working.
